```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('walmart-sales-dataset-of-45stores.csv')

# Plotting histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Weekly_Sales', bins=20, kde=True)
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
holiday_counts = df['Holiday_Flag'].value_counts()
sns.barplot(x=holiday_counts.index, y=holiday_counts.values)
plt.title('Number of Weeks with and without Holiday Events')
plt.xlabel('Holiday Flag')
plt.ylabel('Frequency')
plt.xticks(ticks=[0, 1], labels=['No Holiday', 'Holiday'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
holiday_labels = ['No Holiday', 'Holiday']
plt.pie(holiday_counts, labels=holiday_labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Weeks with and without Holiday Events')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

# Load the dataset
df = pd.read_csv('walmart-sales-dataset-of-45stores.csv')

# Remove non-numeric columns for correlation analysis
numeric_df = df.drop(columns=['Date'])

# Plotting Confusion Matrix (Dummy Example)
plt.figure(figsize=(8, 6))
sns.heatmap([[100, 20], [30, 50]], annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()

# Plotting Heatmap of Correlation Matrix
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Plotting Corner Plot
scatter_matrix(df, alpha=0.2, figsize=(10, 10), diagonal='kde')
plt.suptitle('Corner Plot')
plt.show()

# Plotting Box Plot of Weekly Sales
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Weekly_Sales')
plt.title('Box Plot of Weekly Sales')
plt.ylabel('Weekly Sales')
plt.show()

# Plotting Violin Plot of Weekly Sales
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, y='Weekly_Sales')
plt.title('Violin Plot of Weekly Sales')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset
df = pd.read_csv('walmart-sales-dataset-of-45stores.csv')

# Select relevant features for clustering
X = df[['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]

# Elbow Plot
plt.figure(figsize=(10, 6))
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Set n_init=10
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# Silhouette Plot
plt.figure(figsize=(10, 6))
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Set n_init=10
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
